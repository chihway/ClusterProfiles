```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from astropy.cosmology import FlatLambdaCDM
cosmo = FlatLambdaCDM(H0=70, Om0=0.3)
```

```python
# calculate beam size
#Planck beam = 7 arcmin. At z=0.05, D_A = 200 Mpc with h=0.7 -> Beam size:
xx = cosmo.comoving_transverse_distance(0.05)
beam = 7*np.pi/180.0/60*xx
print(xx,beam)
```

```python
zz=np.arange(100)/1000
dd=cosmo.comoving_transverse_distance(zz)
plt.plot(zz,7*np.pi/180.0/60*dd)
#plot Planck beam vs redshift
```
